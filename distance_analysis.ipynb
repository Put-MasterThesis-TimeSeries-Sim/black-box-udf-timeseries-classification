```python
import pandas as pd
```

```python
distance_df = pd.read_csv("comparison_distance.csv")
```

## CPU - not standardised and not smoothed

```python
filter1 = distance_df["dataType"]=="CPU"
filter2 = distance_df["is_standardised"]==0
filter3 = distance_df["is_smoothed"]==0

filtered = distance_df[filter1 & filter2 & filter3]
filtered
```

## CPU - standardised but not smoothed

```python
filter1 = distance_df["dataType"]=="CPU"
filter2 = distance_df["is_standardised"]==1
filter3 = distance_df["is_smoothed"]==0

filtered = distance_df[filter1 & filter2 & filter3]
filtered
```

## CPU - smoothed but not standardised

```python
filter1 = distance_df["dataType"]=="CPU"
filter2 = distance_df["is_standardised"]==0
filter3 = distance_df["is_smoothed"]==1

filtered = distance_df[filter1 & filter2 & filter3]
filtered
```

## CPU - standardised and smoothed

```python
filter1 = distance_df["dataType"]=="CPU"
filter2 = distance_df["is_standardised"]==1
filter3 = distance_df["is_smoothed"]==1

filtered = distance_df[filter1 & filter2 & filter3]
filtered

```
