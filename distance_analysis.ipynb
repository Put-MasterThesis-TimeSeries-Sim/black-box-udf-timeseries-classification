```python
import pandas as pd
```

```python
distance_df = pd.read_csv("comparison_distance.csv")
```

```python
def rowStyle(row):
    if row.function1 != row.function2:
        return ['background-color: yellow'] * len(row)
    return [''] * len(row)
```

## CPU - not normalized and not smoothed

```python
filter1 = distance_df["dataType"]=="CPU"
filter2 = distance_df["is_normalized"]==0
filter3 = distance_df["is_smoothed"]==0

filtered = distance_df[filter1 & filter2 & filter3]
filtered.sort_values(by=['dtw-python-distance'], inplace=True)
filtered.style.apply(rowStyle,axis = 1)
```

## CPU - normalized but not smoothed

```python
filter1 = distance_df["dataType"]=="CPU"
filter2 = distance_df["is_normalized"]==1
filter3 = distance_df["is_smoothed"]==0

filtered = distance_df[filter1 & filter2 & filter3]
filtered.sort_values(by=['dtw-python-distance'], inplace=True)
filtered.style.apply(rowStyle,axis = 1)
```

```python
filter1 = distance_df["dataType"]=="CPU"
filter2 = distance_df["is_normalized"]==0
filter3 = distance_df["is_smoothed"]==1

filtered = distance_df[filter1 & filter2 & filter3]
filtered.sort_values(by=['dtw-python-distance'], inplace=True)
filtered.style.apply(rowStyle,axis = 1)
```
